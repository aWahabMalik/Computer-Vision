# Importing All usefull Libraries

```python
import numpy as np
import matplotlib.pylab as plt
import cv2 as cv
```

# Reading Image

```python
#Image Reading
im = plt.imread("Monkey.jpg")
#to display the images
plt.imshow(im)
```

# Creating ndArray for grayScale

```python
grayScale = np.zeros( (im.shape[0], im.shape[1]) , dtype=np.uint8)
```

# Converting to GrayScale

```python
for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        grayScale[i,j] = int(im[i,j].mean())
        
plt.imshow(grayScale,cmap = 'gray')
```
